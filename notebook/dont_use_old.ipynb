```python
import geopandas as gpd
from shapely.geometry import Polygon

# 1️⃣ Define your block corners in order (lon, lat) or (x, y)
# Example: a simple rectangular block
block_coords = [
    (-112.25, 33.25),  # point 1
    (-112.24, 33.25),  # point 2
    (-112.24, 33.24),  # point 3
    (-112.25, 33.24),  # point 4
    (-112.25, 33.25)   # back to point 1 to close (optional but nice)
]

block_poly = Polygon(block_coords)

# 2️⃣ Put it into a GeoDataFrame
block_gdf = gpd.GeoDataFrame(
    {"name": ["block_1"]},
    geometry=[block_poly],
    crs="EPSG:4326"      # WGS84 lat/lon (what you’ve been using in class)
)
```
