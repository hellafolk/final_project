```python
import requests

url = "https://gml.noaa.gov/ccgg/mbl/tmp/co2_GHGreference.1023527514_surface.txt"
out_path = "/Volumes/ct_data/mbl_data_ocean/co2_GHGreference_2024_surface_raw.txt"

resp = requests.get(url)
resp.raise_for_status()  # will error if something went wrong

with open(out_path, "wb") as f:
    f.write(resp.content)

print("Saved raw MBL file to:", out_path)

```
